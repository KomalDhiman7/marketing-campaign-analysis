**Load and Clean**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("marketing_campaign_dataset.csv")
```

```python
print(df.head())
```

```python
# remove missing values
df.dropna(inplace=True)

# remove duplicates
df.drop_duplicates(inplace=True)
```

```python
print(df.describe())
```

**Analysis**

***1. Campaign Type vs Conversion Rate***

```python
df.groupby('Campaign_Type')['Conversion_Rate'].mean().plot(kind='bar')
plt.title("Campaign Type vs Conversion Rate")
plt.ylabel("Conversion Rate")
plt.show()
```

***2. Channel Performance***

```python
df.groupby('Channel_Used')['Conversion_Rate'].mean().plot(kind='bar')
plt.title("Channel Used vs Conversion Rate")
plt.show()
```

***3. Engagement vs Conversion***

```python
plt.scatter(df['Engagement_Score'], df['Conversion_Rate'])
plt.title("Engagement Score vs Conversion Rate")
plt.xlabel("Engagement Score")
plt.ylabel("Conversion Rate")
plt.show()
```

***4. ROI Analysis***

```python
df.groupby('Campaign_Type')['ROI'].mean().plot(kind='bar')
plt.title("Campaign Type vs ROI")
plt.show()
```

***5. Customer Segment Analysis***

```python
df.groupby('Customer_Segment')['Conversion_Rate'].mean().plot(kind='bar')
plt.title("Customer Segment vs Conversion Rate")
plt.show()
```

***6. Cost vs Conversion***

```python
plt.scatter(df['Acquisition_Cost'], df['Conversion_Rate'])
plt.title("Acquisition Cost vs Conversion Rate")
plt.xlabel("Cost")
plt.ylabel("Conversion Rate")
plt.show()
```
